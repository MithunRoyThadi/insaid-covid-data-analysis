# **8. Exploratory Data Analysis**
---

```python
covid_data.head()
```

```python
covid_data.set_index('State/UTs').TotalCases.sort_values(ascending=False)[:10].plot(kind='bar')

plt.title("States/UTs which registered highest covid cases", size = 10, color = 'Black')
plt.xlabel('State/UT', size = 12, color = 'blue')
plt.ylabel('Total Cases', size = 12, color = 'red')
```

```python
covid_data.set_index('State/UTs').TotalCases.sort_values(ascending=True)[:10].plot(kind='bar')

plt.title("States/UTs which registered least covid cases", size = 10, color = 'Black')
plt.xlabel('State/UT', size = 10, color = 'blue')
plt.ylabel('Total Cases', size = 10, color = 'red')
```

```python
figure = plt.figure(figsize=[10,10])
covid_data.set_index('State/UTs').TotalCases.sort_values(ascending=False)[:36].plot(kind='pie', explode =  np.ones(36)/36, fontsize=12, 
                                        startangle=0, cmap='inferno', wedgeprops=dict(width=0.15), legend=True)


```

```python
covid_data.set_index('State/UTs').TotalCases.sort_values(ascending=False)[:36].plot(kind='pie', cmap = 'inferno', startangle=0,
                                           fontsize=15,  explode=np.ones(36)/20,
                                           figsize=(20,20),
                                            wedgeprops=dict(width=0.15),legend=True
                                         )
```
